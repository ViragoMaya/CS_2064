### Python Class Introduction
> `A Python class is a blueprint for creating objects. It defines the attributes (data/ sumn that describe it ) and methods (functions) that describe how the object behaves.`
>
> Think of a class like the design for a bicycle. The design tells you that every bicycle has wheels, a frame, pedals, and brakes. But the actual bicycle you ride â€” with its specific color, brand, and speed â€” is an object created from that class.
>
> Each object made from a class can have different values for its attributes, but they all follow the same structure defined by the class.`

### Python Class Definition

> A class in Python is a user-defined data structure. It encapsulates data (attributes) and behavior (methods) into a single unit. Classes support the principles of object-oriented programming (OOP), including encapsulation, inheritance, and polymorphism.
> 
> In Python, a class is defined using the class keyword, and it can include:
>   - Attributes: Variables that hold the state of an object.
>   - Methods: Functions that define the behavior of an object.

### Python Class Example

```python
# Define a Bicycle class
class Bicycle: #everysingle method in you class must have self as the first parameter #every time you refer to attribute you have to use self, self refers to object. Class.attribute#dif memory locations
    def __init__(self, brand, color):
        self.brand = brand #self refers to the instance being created
        self.color = color
        self.speed = 0  # starts at rest

    def pedal(self):
        self.speed += 1
        print(f"{self.brand} bicycle is pedaling. Speed : {self.speed} mph.")

    def brake(self):
        if self.speed > 0:
            self.speed -= 1
            print(f"{self.brand} bicycle is braking. Speed : {self.speed} mph.")
        else:
            print(f"{self.brand} bicycle is already stopped.") #prevents negative speed #defining the class is your blueprint

    def ring_bell(self):
        print(f"{self.brand} bicycle goes 'squeak!' ðŸ””")

# Create a Bicycle object
my_bike = Bicycle("Cannondale", "matte blue")

# Use the Bicycle methods
my_bike.pedal() #creates variables
my_bike.pedal()
my_bike.ring_bell()
my_bike.brake()
my_bike.brake()
my_bike.brake()

```
