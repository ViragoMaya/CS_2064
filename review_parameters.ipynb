# Parameter ðŸ†š Arugment
> - Parameter: A variable listed inside the parentheses in a function definition.
> - Argument: The actual value that is passed int the function with it is called.

```python
# function definition
def greet(name): # name is the parameter
    print(f"Hello {name}")

# function call
greet("Hokie") # "Hokie" is the argument
```

```python
# function definition
# name and role are both parameters
# role has a default value set to "Student"
def introduce(name, role="Student"): 
    print(f"{name} is a {role}.")

# function call
# "Hokie" is the argument
introduce("Hokie") 

# function call
# "Anna Sthesia" and "CRNA" are both arguments
introduce("Anna Sthesia", "CRNA") 
```
