```python
class WeatherReport:
    def __init__(self, location, temperature_c, cloud_percent, condition):
        # location is expected to be a dictionary
        # location example: {"city": "Blacksburg", "state": "Virginia", "country": "US"}
        #below are instance attributes
        self.location = location
        self.temperature_c = temperature_c
        self.cloud_percent = cloud_percent
        self.condition = condition

    def is_cloudy(self): # why do we put self in the parentheses when defining a method? 
        return self.cloud_percent > 40

#method that returns value below
    def is_uncomfortable_weather(self):
        if self.temperature_c < 0:
            return True
        elif self.temperature_c > 30 and self.cloud_percent > 25:
            return True
        return False


```
