# What is a Python set?
>- A Python set is an unordered collection of unique elements.

#### Key Characteristics
>- Unordered: The items have no defined order.
>- Mutable: You can add and/or remove elements.
>- No Dubplicates: Each element is unique.
>- Set Operations: Union (|), Intersection (&), Difference (-), Symmetric Difference (^)

#### Why use a set?
>- Store items without duplicates.
>- Perform operations like unions, intersections, etc (Vinn Diagrams anyone?). 

```python
# Unordered Set Example
buildings = {"McBryde Hall", "Torgersen", "New Classroom"}
for building in buildings:
    print(building)
#print(buildings[0]) # Error - set object is not subscriptable
```

```python
# Mutable Set Example
buildings = {"McBryde Hall", "Torgersen", "New Classroom"}
buildings.add("Goodwin Hall")
buildings.add("Newman Library")
buildings.remove("New Classroom")

for building in buildings:
    print(building)
```

```python
# No Duplicates Set Example
buildings = {"McBryde Hall", "Torgersen", "New Classroom", "Torgersen"}
buildings.add("Torgersen")
buildings.add("Torgersen")
# we love Torgersen, lets add it more!
for _ in range(100):
    buildings.add("Torgersen")
    
for building in buildings:
    print(building)
```

```python
# Set Operations Example
sue_schedule = {
    ("CS1064", "McBryde Hall"),
    ("MATH2100", "Robeson Hall"),
    ("PHYS1500", "McBryde Hall"),
    ("ENGL1100", "Shanks Hall"),
    ("CHEM1003", "Robeson Hall")
}
flay_schedule = {
    ("CS1064", "McBryde Hall"),
    ("CS1014", "New Classroom"),
    ("PHYS1500", "McBryde Hall"),
    ("ENGL2100", "Shanks Hall"),
    ("MATH2100", "Torgersen"),
}
print(f"Common Courses: {sue_schedule & flay_schedule}")
print(f"Both Schedules: {sue_schedule | flay_schedule}")
print(f"Different Courses: {sue_schedule - flay_schedule}")
print(f"Different Courses: {flay_schedule - sue_schedule}")
print(f"Symmetrical Difference (not in intersection):\n {sue_schedule ^ flay_schedule}")
```
