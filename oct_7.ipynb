### Abstract Classes
>- An abstract class in Python is a blueprint for other classes. It defines a common interface and structure that its subclasses must follow.
>- Think of it as a template that outlines what a class should do, but leaves the details to be filled in by subclasses.

**Why use abstract classes?**
>- Abstract classes are used when you want to ensure that certain methods are implemented in all derived classes, without providing a complete implementation in the base class.

**Key Characteristics of Abstract Classes**
>- **Cannot** Be Instantiated
>- Defined Using the abc Module
>- Conatins Abstract Methods
>- Used to Enforce a Contract
>- Supports Polymorphism
