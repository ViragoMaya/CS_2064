### Python Classes
- `__init__()`
<br>This is the constructor method in Python. It runs automatically when a new object of the class is created. It’s typically used to initialize the object’s attributes.
- `__str__()`
<br>This method defines the string representation of an object. When you use print() on an object, Python calls `__str__()` to determine what to display.
- `Testing Class`
<br>This refers to the practice of writing code to verify that a class behaves as expected. It often involves creating instances and checking their attributes or methods using assertions.

```python
class WeatherReport:#camel case 
    def __init__(self, location, temperature_c, humidity_percent, condition):
        # location is expected to be a dictionary
        #wtv method you have in ur class self must be the first parameter (varibles)
        self.location = location  # self.location is teh class, and location is the parameter -- 2 dif variables living in two dif spots in ur memory.
        self.temperature_c = temperature_c
        self.humidity_percent = humidity_percent
        self.condition = condition
        #initialize attributes with the provided parameters

    def __str__(self):
        loc_parts = [self.location.get(key) for key in ['city', 'state', 'country'] if key in self.location]
        loc_str = ", ".join(loc_parts)
        return (f"Weather Report for {loc_str}:\n"
                f"  Temperature: {self.temperature_c}°C\n"
                f"  Humidity: {self.humidity_percent}%\n"
                f"  Condition: {self.condition}")
```

```python
import unittest

class TestWeather(unittest.TestCase):
    def setUp(self):
        # setUp() will run before each test method - defualt values
        location = {'city': 'Blacksburg', 'state': 'VA', 'country': 'US'}
        self.weather = WeatherReport(location, 0, 30, 'Sunny')
    
    def test_temperature_f_0(self): # test method names must start with 'test' # temp is the method we are testing # 0 lets us know we r testing that value
        # step 1: Set up -- what u start with each and every test method, you start over.
          # done in our setUp()

        # step 2: Call the method
        result = self.weather.temperature_f()

        # step 3: Define expected value
        expected_value = 32

        # step 4: Assert
        self.assertEqual(result, expected_value) # these two values should be equal

    def test_temperature_f_100(self):
        self.weather.temperature_c = 100
        result = self.weather.temperature_f()
        expected_value = 212
        self.assertEqual(result, expected_value)

    def test_is_humid(self):
        result = self.weather.is_humid()
        #expected = False
        self.assertFalse(result) # can also js assertrue that way you dont have to define expected

    # continue testing the WeatherReport class
    def test_is_humid_true(self):
        self.weather.humidity_percent = 80 # changed set up so that it returns true
        result = self.weather.is_humid() # call the method '()'
        self.assertTrue(result) #result is now testes

    # Run the tests
    def test_weather_alert_storm(self):
        cond = ["Storm", "Heavy Rain", "Thunderstorm"] # change the condition to storm bc set up is sunny # this tests if part 
        for c in cond:
            self.weather.condition = c # change the condition to storm
            result = self.weather.weather_alert() # call the method
            expected_value = "Storm Warning"
            self.assertEqual(result, expected_value)

    # testing teh elif branch (the true then the  false)

    def test_weather_alert_storm_true(self):
        self.weather.temperature_c = 40 # testing edge cases using 35 
        result = self.weather.weather_alert()
        expected_value = "Heat Alert"
        self.assertEqual(result, expected_value)




    def test_is_uncomfortable_weather_if(self):
        self.weather.temperature_c = -1 # we expect it to return true
        result = self.weather.is_uncomfortable_weather()
        self.assertTrue(result)
    
    def test_is_uncomfortable_weather_if_elif_tt(self):
        self.weather.temperature_c = 40
        self.weather.humidity_percent = 80
        result = self.weather.is_uncomfortable_weather()
        self.assertTrue(result)
    
    def test_is_uncomfortable_weather_else(self):
        self.weather.temperature_c = 20
        self.weather.humidity_percent = 50 # both conditions are false so we expect it to return false -- or like less -- numbers lowk alrd in set up tho?
        result = self.weather.is_uncomfortable_weather()
        self.assertFalse(result)




unittest.main(argv=[''], verbosity=2, exit=False) # dont need to have any of these things, these are all default but verbosity tells you which tests acc pass. 
```
