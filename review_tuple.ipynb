# What is a Python Tuple?
>- A Python tuple is an ordered, immutable collection of elements. 
>- Tuples can hold items of any data type, and are defined by placing elements inside parentheses () and separating them with commas.

### Key Characteristics:
>- Ordered 
>    - Tuples maintain the order of elements.
>- Immutable
>    - Tuples cannot be changed after creation.
>- Indexable
>    - You can access tuple elements by index, and use slicing.
>- Heterogeneous
>    - Tuples can contain elements of different types, just like lists.

```python
# Example of Ordered and Heterogeneous
student = ("Heath Hillman", 42, "Physics", 3.8)

for index, value in enumerate(student):
    print(f"{index}: {value}")
```

```python
# Example of Immutable
student = ("Heath Hillman", 161, "Physics", True)
#student[0] = "Sue Flay" # Error
#student[4] = "Error" # Error
#student.add("Error") # Error
```

```python
# Example of Indexable
student = ("Heath Hillman", 161, "Physics", True)

print(f"{student[0]} is majoring in {student[2]}.")
```

```python
# Tuples can be keys for a Dictionary
locations = {
    (37.2, -80.4): "Blacksburg, VA",
    (38.9, -77.0): "Washington, DC",
    (34.0, -118.2): "Los Angeles, CA"
}

# Accessing a value using a tuple key
city = locations[(37.2, -80.4)]
print(city)  # Output: Blacksburg, VA
```
